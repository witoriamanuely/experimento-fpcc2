## Código Auxiliar

`````python

def plot_compressed_ratio(compressed_ratios_lzw, compressed_ratios_huffman, compressed_ratios_flc, compressed_ratios_hflc):
    # Criar figura e eixos
    fig, ax = plt.subplots(figsize=(10, 6))


    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_lzw['file'], compressed_ratios_lzw['compressed_ratio'], marker='o', label='LZW')

    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_huffman['file'], compressed_ratios_huffman['compressed_ratio'], marker='o', label='Huffman')

    # Plotar os dados de compressed_ratios_flc_large
    ax.plot(compressed_ratios_flc['file'], compressed_ratios_flc['compressed_ratio'], marker='o', label='FLC')

    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_hflc['file'], compressed_ratios_hflc['compressed_ratio'], marker='o', label='HFLC')


    # Rotacionar os rótulos do eixo x, se necessário
    plt.xticks(rotation=45)

    # Definir os rótulos dos eixos e o título do gráfico
    ax.set_xlabel('Arquivo')
    ax.set_ylabel('Taxa de Compressão')
    ax.set_title('Progressão da Taxa de Compressão')

    # Exibir uma legenda no gráfico
    ax.legend()

    # Exibir o gráfico
    plt.grid(True)
    plt.show()
`````
