## Código Auxiliar

```python

def plot_compressed_metrics(compressed_ratios_lzw, compressed_ratios_huffman, compressed_ratios_flc, compressed_ratios_hflc, key, title):
    
    # Criar figura e eixos
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_lzw.index, compressed_ratios_lzw[key], marker='o', label='LZW', color='purple')

    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_huffman.index, compressed_ratios_huffman[key], marker='o', label='Huffman', color='blue')

    # Plotar os dados de compressed_ratios_flc_large
    ax.plot(compressed_ratios_flc.index, compressed_ratios_flc[key], marker='o', label='FLC', color='red')

    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_hflc.index, compressed_ratios_hflc[key], marker='o', label='HFLC', color='green')

    # Rotacionar os rótulos do eixo x, se necessário
    plt.xticks(rotation=45)

    # Definir os rótulos dos eixos e o título do gráfico
    ax.set_xlabel('Arquivo')
    ax.set_ylabel(title)
    ax.set_title(title)

    # Exibir uma legenda no gráfico
    ax.legend()

    # Exibir o gráfico
    plt.grid(True)
    plt.show()



def normalized(value, min_value, max_value):
    return ((value - min_value) / (max_value - min_value)) * 100
```

```python
def plot_compressed_metrics_offset(compressed_ratios_lzw, compressed_ratios_huffman, compressed_ratios_flc, compressed_ratios_hflc, key, title):
    # Criar figura e eixos
    fig, ax = plt.subplots(figsize=(10, 6))

    # Deslocamento vertical para separar as linhas
    offset = 0.05

    # Plotar os dados de compressed_ratios_lzw_large
    ax.plot(compressed_ratios_lzw.index, compressed_ratios_lzw[key], marker='o', label='LZW', color='purple')

    # Plotar os dados de compressed_ratios_huffman_large com deslocamento vertical
    ax.plot(compressed_ratios_huffman.index, compressed_ratios_huffman[key] + offset, marker='o', label='Huffman', color='blue')

    # Plotar os dados de compressed_ratios_flc_large com deslocamento vertical
    ax.plot(compressed_ratios_flc.index, compressed_ratios_flc[key] + 2 * offset, marker='o', label='FLC', color='red')

    # Plotar os dados de compressed_ratios_hflc_large com deslocamento vertical
    ax.plot(compressed_ratios_hflc.index, compressed_ratios_hflc[key] + 3 * offset, marker='o', label='HFLC', color='green')

    # Rotacionar os rótulos do eixo x, se necessário
    plt.xticks(rotation=45)

    # Definir os rótulos dos eixos e o título do gráfico
    ax.set_xlabel('Arquivo')
    ax.set_ylabel(title)
    ax.set_title(title)

    # Exibir uma legenda no gráfico
    ax.legend()

    # Exibir o gráfico
    plt.grid(True)
    plt.show()
```
