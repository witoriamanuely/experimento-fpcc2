## Código Auxiliar

```python

def plot_compressed_metrics(compressed_ratios_lzw, compressed_ratios_huffman, compressed_ratios_flc, compressed_ratios_hflc, key, title):
    # Criar figura e eixos
    fig, ax = plt.subplots(figsize=(10, 6))


    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_lzw.index, compressed_ratios_lzw[key], marker='o', label='LZW')

    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_huffman.index, compressed_ratios_huffman[key], marker='o', label='Huffman')

    # Plotar os dados de compressed_ratios_flc_large
    ax.plot(compressed_ratios_flc.index, compressed_ratios_flc[key], marker='o', label='FLC')

    # Plotar os dados de compressed_ratios_hflc_large
    ax.plot(compressed_ratios_hflc.index, compressed_ratios_hflc[key], marker='o', label='HFLC')


    # Rotacionar os rótulos do eixo x, se necessário
    plt.xticks(rotation=45)

    # Definir os rótulos dos eixos e o título do gráfico
    ax.set_xlabel('Arquivo')
    ax.set_ylabel(title)
    ax.set_title(title)

    # Exibir uma legenda no gráfico
    ax.legend()

    # Exibir o gráfico
    plt.grid(True)
    plt.show()

    


def normalized(value, min_value, max_value):
    return ((value - min_value) / (max_value - min_value)) * 100
```
